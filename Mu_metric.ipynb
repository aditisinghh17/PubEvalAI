```python
import openai

# Set up OpenAI API key
openai.api_key = ''

# Define a function to classify queries
def classify_query_gpt(responses):
    prompt = f"Classify the following responses of the chatbot in a domain: {responses}"

    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",  # Specify the model
        messages=[{"role": "user", "content": prompt}], # Include the messages parameter
        max_tokens=10,
        temperature=0.0
    )

    domain = response.choices[0].message.content.strip() # Access content from message object
    return domain

response = df["Response"].tolist()
predicted_domains = []
print(response)
for i in response:
    predicted_domain = classify_query_gpt(i)
    predicted_domains.append(predicted_domain)

# Add the predicted domains to the DataFrame
# df["predicted_domain"] = predicted_domains
# # Example usage
# query =  "who is my brother"
# predicted_domain = classify_query_gpt(query)
# print(predicted_domain)
```

```python

df['Domain'] = predicted_domains
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define your list of texts
texts = predicted_domains

# Generate embeddings
embeddings_domain = model.encode(texts)

print(embeddings_domain)
```

```python
intended_domain = ["Mental Health"]
embeddings_intended = model.encode(intended_domain)
similarities = util.pytorch_cos_sim(embeddings_domain, embeddings_intended)
print(similarities)
```

```python
print(similarities.shape)
```

```python
similarity_threshold = 0.5
results = []
for i, query in enumerate(predicted_domains):
    similarity_score = similarities[i][0].item()
    if similarity_score < similarity_threshold:
        results.append(("Irrelevant"))
    else:
        results.append(("Relevant"))

# Output results
for  status in results:
    print({status})
```

```python
df['Relevance'] = results
```

```python
Correctly_handled_irrelevant_queries = df[df['Relevance'] == "Irrelevant"].shape[0]
total_irreleavent_queries = df.shape[0]
print(total_irreleavent_queries)
print(Correctly_handled_irrelevant_queries)
# Calculate the MU Score
if total_irreleavent_queries > 0:
    mu_score = (Correctly_handled_irrelevant_queries / total_irreleavent_queries) * 100
else:
    mu_score = 0  # Handle the case where there are no queries

# Output the MU Score
print(f"MU Score: {mu_score:.2f}%")
```
